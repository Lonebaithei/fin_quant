```python
%matplotlib inline
import quantstats as qs

# extend pandas functionality with metrics, etc.
qs.extend_pandas()

# list of tickers to fetch returns for
tickers = ["NVDA","TSM","LMT","TSM","ADV","AIRI","CAT","COTY",'DIS','ERIC','GNPX','INTC','MU','NOC','RTX','SMCI','STEM','T','TDOC','UGI','UNH','VNQ']

# download returns into a dict: {ticker: returns_series}
stocks = {t: qs.utils.download_returns(t) for t in tickers}

# show sharpe ratio for each ticker
for t, s in stocks.items():
	print(t, qs.stats.sharpe(s))

# or using extend_pandas() on a single series, e.g. NVDA
stocks['NVDA'].sharpe()
```
